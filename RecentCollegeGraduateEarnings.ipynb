### Summary of scatter plots

The graphs show that there is a high correlation between those that have graduated with a certain degree/major and the number that are employed.  However, there is not a strong relationship between the number that have a certain major degree and the median salary or unemployment rate.  This makes sense from the perspective that median salary should be determined by the supply and demand relationships for certain skills.  Nevertheless there is a strong relationship between the median salary and the share of women that have a particular degree.  This occurs because women tend to select themselves into less lucrative fields because of various reasons related to preferences, work-life balance and sometimes because of gender biases that made it seemingly harder to enter more male-dominated and lucrative majors that would pay-off in terms of higher salaries.

### Next steps
Use a grouped bar plot to compare the number of men with the number of women in each category of major.
Use a box plot to explore the distributions of median salaries and unemployment rates.
Use a hexagonal bin plot to visualize the columns that had dense scatter plots from earlier in the project.

```python
#Grouped bar plot to compare the number of men with the number of women in each category of majors

import matplotlib.pyplot as plt

#sum over the major categories to get aggregate rates
agg_major_category = recent_grads.groupby(by=["Major_category"]).sum()
#print(agg_major_category)
#Format some of the major category names
agg_major_category["fname"] = agg_major_category.index

mapping_dict = {"Agriculture & Natural Resources":"Ag & Nat. Sci.",
                "Biology & Life Science":"Bio & Life Sci.", 
                "Humanities & Liberal Arts":"Humanities & Lib. Arts",
               "Industrial Arts & Consumer Services":"Ind. Arts & Cons. Svc.",
                "Psychology & Social Work":"Psych. & Soc. Work"
               }
agg_major_category["fname"]=agg_major_category["fname"].replace(mapping_dict)

pos = list(range(len(agg_major_category['Women'])))
width = 0.25

# Plotting the bars.  Create figure size according to the number of major categories
fig, ax = plt.subplots(figsize=(len(agg_major_category)+len(agg_major_category)*width,5))

# Create a bar with pre_score data,
# in position pos,
plt.bar(pos, agg_major_category['Women'], width, alpha=0.20, color='#EE3224', label="Women")

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos],agg_major_category['Men'], width, alpha=0.5, color='#F78F1E',label="Men")

# Set the labels and axis ticks
ax.set_ylabel('Number')
ax.set_title('Number in Major Category by Gender')
ax.set_xticks(pos)
ax.set_xticklabels(agg_major_category["fname"], rotation=45)
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(agg_major_category['Women'] + agg_major_category['Men'])] )

# Adding the legend and showing the plot
plt.legend(['Women', 'Men'], loc='upper left')
plt.grid()
plt.show()
```
